```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_anomaly = pd.read_csv(r"C:\Users\Home\Documents\REZEL\program_internship_paul_wurth\data\df_1min_2025-02-24_2025-03-24.csv")
df_reconstruction = pd.read_csv(r"C:\Users\Home\Documents\REZEL\program_internship_paul_wurth\data\data_rec.csv")
print(df_anomaly.head())
print(df_reconstruction.head())
```

```python
MSE = (df_anomaly - df_reconstruction) ** 2
plt.figure(figsize=(12, 6))
sns.lineplot(data=MSE)
plt.title("Mean Squared Error between Anomaly and Reconstruction")
plt.xlabel("Time")
plt.ylabel("MSE")
plt.show()
```

```python
#here plot each column anomaly vs reconstruction in one plt for each column
for column in df_anomaly.columns:
    plt.figure(figsize=(12, 6))
    sns.lineplot(data=df_anomaly[column], label='Anomaly')
    sns.lineplot(data=df_reconstruction[column], label='Reconstruction')
    plt.title(f"Anomaly vs Reconstruction for {column}")
    plt.xlabel("Time")
    plt.ylabel("Value")
    plt.legend()
    plt.show()
```
